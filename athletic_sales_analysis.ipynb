```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_2020_path = Path('Resources/athletic_sales_2020.csv')
athletic_2021_path = Path('Resources/athletic_sales_2021.csv')

athletic_2020_df = pd.read_csv(athletic_2020_path)
athletic_2021_df = pd.read_csv(athletic_2021_path)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_2021_df.columns

combine_athletic_20_21 = pd.concat([athletic_2020_df , athletic_2021_df], axis="rows", join = "inner")
combine_athletic_index = combine_athletic_20_21.reset_index()
combine_athletic_index.head(20)
```

```python
# Check if any values are null.
null_check = combine_athletic_index.isnull().sum()
null_check
```

```python
# Check the data type of each column
combine_athletic_index.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combine_athletic_index ['invoice_date'] = pd.to_datetime(combine_athletic_index ['invoice_date'], format="%m/%d/%y")
```

```python
# Confirm that the "invoice_date" data type has been changed.

combine_athletic_index ['invoice_date'].dtype
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

combine_athletic_metrics = combine_athletic_index.groupby(['region', 'state', 'city']) \
    .agg({'units_sold': 'sum'}).reset_index()
combine_athletic_metrics.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
combine_athletic_metrics_sorted = combine_athletic_metrics.sort_values(by='Total_Products_Sold', ascending=False)

combine_athletic_metrics_sorted.head(5)     


# Show the top 5 results.

 
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

product_location = pd.pivot_table(combine_athletic_index,
                                  index=['region', 'state', 'city'],
                                  values='units_sold',
                                  aggfunc = 'sum',
                                  sort=False).reset_index()  



# Rename the "units_sold" column to "Total_Products_Sold"

product_location.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.

product_location_sort = product_location.sort_values(by='Total_Products_Sold', ascending=False)
product_location_sort.head(5)   
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

combine_athletic_index.groupby(['retailer','region','state','city'])[['total_sales']].sum().sort_values(by='total_sales', ascending=False).rename(columns={'total_sales':'Total Sales'}).head()
# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.


sales_region = pd.pivot_table(combine_athletic_index,
                                  index=['region', 'state', 'city'],
                                  values='total_sales',
                                  aggfunc = 'sum',
                                  sort=False).reset_index()

sales_region.rename(columns={'total_sales': 'Total Sales'}, inplace=True)


sales_region_sort = sales_region.sort_values(by='Total Sales', ascending=False)
sales_region_sort.head(5)   
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"

total_sales_region = combine_athletic_index.groupby(['retailer' ,'region', 'state', 'city']) \
    .agg({'total_sales': 'sum'}).reset_index()

total_sales_region.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
total_sales_region_sorted = total_sales_region.sort_values(by= ['Total Sales'], ascending=[False])
total_sales_region_sorted.groupby('retailer')  #doing a second groupby to group results by retailer  
total_sales_region_sorted.head(5) 
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

# Optional: Rename the "total_sales" column to "Total Sales" # Show the top 5 results.

sales_retail_total = pd.pivot_table(combine_athletic_index,
                                  index=['retailer' , 'region', 'state', 'city'],
                                  values= 'total_sales',
                                  aggfunc = 'sum' ,
                                  sort=False).reset_index()


sales_retail_total.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

sales_retail_tot = sales_retail_total.sort_values(by=['Total Sales'], ascending=[False]).groupby('retailer').head(5) 
sales_retail_tot.head(5) 
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.


womens_athletic_df = combine_athletic_index [combine_athletic_index['product'].str.contains \
                                         ("Women's Athletic Footwear", case=False)].reset_index()

womens_athletic_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.

womens_athletic_sold_df = womens_athletic_df.groupby(['retailer' ,'region', 'state', 'city']) \
    .agg({'units_sold': 'sum'}).reset_index()

womens_athletic_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

womens_sold_df_sorted = womens_athletic_sold_df.sort_values(by= ['Womens_Footwear_Units_Sold'], ascending=[False])
womens_sold_df_sorted.groupby('retailer')   
womens_sold_df_sorted.head(5) 
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.

womens_sold_df = pd.pivot_table(womens_athletic_df,
                                  index=['retailer' , 'region', 'state', 'city'],
                                  values= 'units_sold',
                                  aggfunc = 'sum' ,
                                  sort=False).reset_index() 

womens_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

womens_sold_tot = womens_sold_df.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=[False]).groupby('retailer').head(5) 
womens_sold_tot.head(5) 
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"

# Show the table.

womens_date_df = pd.pivot_table(womens_athletic_df,
                                  index=['invoice_date'],
                                  values= 'total_sales',
                                  aggfunc = 'sum' ,
                                  sort=False).reset_index() 


womens_date_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

womens_date_df['invoice_date'] = pd.to_datetime(womens_date_df['invoice_date'], format="%m/%d/%y")
womens_date_df.set_index('invoice_date', inplace=True)

womens_date_df.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".

womens_date_day = womens_date_df.resample('D').sum()

womens_date_day_resample = womens_date_day.sort_values(by=['Total Sales'], ascending=[False]).head(10) 
womens_date_day_resample.head(10) 
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".

womens_date_week = womens_date_df.resample('W').sum()

womens_date_week_resample = womens_date_week.sort_values(by=['Total Sales'], ascending=[False]).head(10) 
womens_date_week_resample.head(10) 
```
